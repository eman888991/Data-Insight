## <span style = "color:mediumorchid;">Imputing missing values
### <span style = "color:violet;">  Why do you need to fill in the missing data?<span><span style = "color:dimgray;"> Because most of the machine learning models that you want to use will provide an error if you pass NaN values into it. The easiest way is to just fill them up with 0, but this can reduce your model accuracy significantly.<span>
Missing values are usually represented in the form of <span style = "color:orange;"><font size="4">Nan</span> 
or <span style = "color:orange;"><font size="4">null</span> 
or <span style = "color:orange;"><font size="4">None</span> 
in the dataset.

The dataset used for illustration purpose is related campus recruitment and taken from [ Kaggle page on Campus Recruitment](https://www.kaggle.com/benroshan/factors-affecting-campus-placement)

```python
df = pd.read_csv("../input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv")
df.head()
```

### Make a note of<span style = "color:orange;"> NaN</span>  value under the salary column.

#### Different methods that you can use to deal with the missing data.
## The methods I will be discussing are:
### 1- Deleting the columns with missing data <span style = "color:orange;"> df.dropna(axis=1)</span>
### 2- Deleting the rows with missing data <span style = "color:orange;"> df.dropna(axis=0)</span>
### 3- Filling the missing data with (Mean,Median,Mode)value of other salary values or Constant <span style = "color:orange;"> df['salary'].fillna(0.0)</span>

```python
df['salary'] = df['salary'].fillna(0.0)
df.head()
```

## <span style = "color:mediumorchid;">Indexing
we can index by sl_no as it is a serial number more index by status as I care aboute it

```python
df = df.set_index(['sl_no','status'])
df.head()
```

## <span style = "color:mediumorchid;">Apply Function

```python
df['salary'].apply([np.min, np.max, np.mean])
```
